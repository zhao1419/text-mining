```python
title_list = pd.DataFrame(titles)
body_list = pd.DataFrame(body)
comments_list = pd.DataFrame(comments)
comments_list
```

```python
df = pd.concat([title_list,body_list],axis = 1)#合并标题和原文
df.columns = ['title_list', 'body_list']
df
```

```python
df = df.applymap(str)#转换str
df
```

```python
df['df_sum'] = df.apply(lambda x: x.sum(), axis=1)#相加标题和原文
df
```

```python
df_comments = pd.concat([comments_list],axis = 1)

type(df_comments)
```

```python
df_comments = df_comments.applymap(str)#转换str
df_comments
```

```python
df_comments['comments_sum'] = df_comments.apply(lambda x: x.sum(), axis=1)# 评论相加
df_comments
```

```python
df_comments.comments_sum
```

```python
df = pd.concat([df.df_sum,df_comments.comments_sum],axis = 1)#合并本文和评论. 在这新建csv，以下新建python
df.head(20)
```

```python
df.to_csv("카폐 크롤링 질문40000.csv", encoding='utf-8', index=False)
```
